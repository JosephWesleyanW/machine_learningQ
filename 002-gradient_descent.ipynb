# Ex 1.0 Simple cost function

   ##                           $$f(x)=x^2+x+1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
f(5)
```

```python
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
x_1[:10]
```

## Slope and Derivatives

```python
def df(x):
    return 2*x +1
```

```python
# Cost and derivative side by side

plt.figure(figsize=[15, 5])

# Cost function's Chart
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3)

# Derivative Chart
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```

# Ex 1.1

## Gradient Descent (slope) Algorithm

```python
x = 3
prevX = 0
stepMultiplier = 0.1
precision = 0.00001

x_list = [x]
slopeList = [df(x)]

for n in range (500):
    prevX = x
    gradient = df(prevX)
    x = prevX - (stepMultiplier * gradient)
    
    stepSize = abs(x - prevX)
    
    x_list.append(x)
    slopeList.append(df(x))
    
    if stepSize < precision:
        print(f'Loop ran {n} times:')
        break
    
print('Local minimum occurs at:              ', x)
print('Slope or df(x) value at this point is: ', df(x))
print('f(x) value or cost at this point is:   ', f(x))
```

```python
plt.figure(figsize=[20, 5])

# Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# Derivative (up close)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
```
